# Pegando  cotações de ações com o Python

- pandas_datareader

### Pegando cotação de 1 ação

```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt

# vamos pegar cotação do Indice e de Empresas do Mercado Financeiro

#indice
df = web.DataReader(f'^BVSP', data_source='yahoo', start=f'01-01-2022', end='25-04-2022')
display(df)
df["Adj Close"].plot(figsize=(15, 10))
plt.show()

#PETR4
df = web.DataReader(f'PETR4.SA', data_source='yahoo', start=f'01-01-2022', end='25-04-2022')
display(df)
df["Adj Close"].plot(figsize=(15, 10))
plt.show()

#VALE3
df = web.DataReader(f'VALE3.SA', data_source='yahoo', start=f'01-01-2022', end='25-04-2022')
display(df)
df["Adj Close"].plot(figsize=(15, 10))
plt.show()

#ITUB4
df = web.DataReader(f'ITUB4.SA', data_source='yahoo', start=f'01-01-2022', end='25-04-2022')
display(df)
df["Adj Close"].plot(figsize=(15, 10))
plt.show()

#USIM5
df = web.DataReader(f'USIM5.SA', data_source='yahoo', start=f'01-01-2022', end='25-04-2022')
display(df)
df["Adj Close"].plot(figsize=(15, 10))
plt.show()
```

### Pegando cotação de uma carteira de ação em um intervalo de datas específico

```python
data_inicial = "01-01-2022"
data_final = "25-04-2022"

empresas_df = pd.read_excel("Empresas.xlsx")
display(empresas_df)

for empresa in empresas_df['Empresas']: 
    print(f"{empresa}:")
    df = web.DataReader(f'{empresa}.SA', data_source='yahoo', start=data_inicial, end=data_final)
    display(df)
    df["Adj Close"].plot(figsize=(15, 10))
    plt.show()
```
